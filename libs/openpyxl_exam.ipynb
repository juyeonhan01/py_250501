```python
from openpyxl import Workbook, load_workbook

# 새로운 엑셀 파일 객체 생성 및 저장
w1 = Workbook()
ws1 = w1.active
w1.save("test.xlsx")

# 기존에 저장한 파일 열기
w2 = load_workbook('test.xlsx')
ws2 = w2.active

# 데이터 쓰기
ws2['A1'] = 'hello'
ws2['C3'] = 'python'

# 여러 셀에 같은 값 쓰기
a = 'python'
for row in ws2['D5':'F8']:
    for cell in row:
        cell.value = a

# 숫자 연속 쓰기
num = 1
for row in ws2["A1":"C3"]:  # ws2로 통일
    for cell in row:
        cell.value = num
        num += 1

# 최종 저장
w2.save("test.xlsx")
```

```python
#엑셀 시트 생성 및 이름 설정

#새로운 시트 생성
newws = w2.create_sheet()

#생성한 시트의 이름 변경
newws.title = '마지막시트'

#원하는 위치(index)
w2.create_sheet("1번시트",0)

w2.save("test.xlsx")
```

```python
#튜플 리스트 차이점

#리스트이름 = [요소 1,2,3,4]

#튜플이름 = [요소 1,2,3,4]

# print(리스트이름[0])
# print(튜플이름[0])

# 리스트는 수정 가능
# 리스트이름[0] = 값

# 튜플은 수정 불가
# 튜플이르메[0]=값


item1 = {'품목':'바나나','가격':'5000','수량':1,'날짜':'2025년 5월 8일'}
item2 = {'품목':'커피;','가격':'3000','수량':2,'날짜':'2025년 6월 9일'}
item3 = {'품목':'책','가격':'14000','수량':3,'날짜':'2025년 7월 10일'}
item4 = {'품목':'펜','가격':'1000','수량':4,'날짜':'2025년 8월 11일'}


items = [item1, item2, item3, item4]

from openpyxl import Workbook
wb = Workbook()
ws = wb.active
ws['A1']='품목'
ws['B1']='가격'
ws['C1']='수량'
ws['D1']='날짜'

i = 2
for item in items:
    ws['A'+str(i)] = item['품목']
    ws['B'+str(i)] = item['가격']
    ws['C'+str(i)] = item['수량']
    ws['D'+str(i)] = item['날짜']
    i += 1

wb.save('test2.xlsx')
```
