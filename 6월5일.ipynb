```python
#Q5

import pandas as pd

# Series 데이터 생성
a = pd.Series([20, 15, 30, 25, 35], name='age')
b = pd.Series([68.5, 60.3, 53.4, 74.1, 80.7], name='weight')
c = pd.Series([180, 165, 155, 178, 185], name='height')

## 1. human DataFrame 만들기
# a, b, c를 열로 갖는 DataFrame 생성
human = pd.DataFrame({'age': a, 'weight': b, 'height': c})
print("1. DataFrame 생성:\n", human, '\n')


def main():
    ### 2. loc(), iloc() 함수를 이용하여 `age`를 추출
    print("2. loc를 이용한 age 추출:\n", human.loc[:, 'age'], '\n')
    print("2. iloc를 이용한 age 추출:\n", human.iloc[:, 0], '\n')
    
    ### 3. loc(), iloc() 함수를 이용하여 `weight`와 `height`만 추출
    print("3. loc를 이용한 weight와 height 추출:\n", human.loc[:, ['weight', 'height']], '\n')
    print("3. iloc를 이용한 weight와 height 추출:\n", human.iloc[:, [1, 2]], '\n')
     
    ### 4. 새로운 데이터 `sex`를 `human`에 추가
    sex = ['F', 'M', 'F', 'M', 'F']
    human['sex'] = sex
    print("4. sex 열 추가 후:\n", human, '\n')
    
    ### 5. `human`에서 `height`를 삭제
    tmp = human.drop(columns='height')
    print("5. height 삭제 후:\n", tmp, '\n')


if __name__ == "__main__":
    main()
```

```python
#Q6

import numpy as np
import pandas as pd

# 2x2, 3x3 무작위 정수 DataFrame 생성
A = pd.DataFrame(np.random.randint(0, 10, (2, 2)))
B = pd.DataFrame(np.random.randint(0, 10, (3, 3)))

print("DataFrame A \n", A)
print("DataFrame B \n", B)

# 모양이 다른 A, B 연산 시 NaN이 생기므로 fill_value 사용
add = A.add(B, fill_value=0)
print("A + B:\n", add, '\n')

sub = A.sub(B, fill_value=0)
print("A - B:\n", sub, '\n')

mul = A.mul(B, fill_value=1)
print("A * B:\n", mul, '\n')

div = A.div(B, fill_value=1)
print("A / B:\n", div, '\n')


# 3x3 DataFrame C 생성
C = pd.DataFrame([[1, 3, 5], [15, 10, 5], [2, 8, 5]],
                 index=['a', 'b', 'c'],
                 columns=['d', 'e', 'f'])

# c 행에 대해 오름차순 정렬
row_C = C.loc['c'].sort_values()

# e 열에 대해 내림차순 정렬
column_C = C.sort_values(by='e', ascending=False)

print("c 행 오름차순 정렬:\n", row_C, '\n')
print("e 열 내림차순 정렬:\n", column_C, '\n')

# CSV 저장 및 불러오기
row_C.to_csv('row_C.csv', index=False)
load_C = pd.read_csv('row_C.csv')

print("CSV 파일 불러오기:\n", load_C)
```
